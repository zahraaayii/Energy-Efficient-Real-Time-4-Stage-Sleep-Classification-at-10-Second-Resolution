```python
from tensorflow.keras import backend as K

def f1(y_true, y_pred):
    def recall_m(y_true, y_pred):
        TP = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        Positives = K.sum(K.round(K.clip(y_true, 0, 1)))

        recall = TP / (Positives+K.epsilon())
        return recall

    def precision_m(y_true, y_pred):
        TP = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        Pred_Positives = K.sum(K.round(K.clip(y_pred, 0, 1)))

        precision = TP / (Pred_Positives+K.epsilon())
        return precision
    precision, recall = precision_m(y_true, y_pred), recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python
from qkeras.utils import _add_supported_quantized_objects
co = {}
_add_supported_quantized_objects(co)
co['f1'] = f1
```

```python
model=load_model(base_dir+"models/cnn_fp8.h5", custom_objects=co)
model.summary()
```

```python
for layer in model.layers:
    if layer.__class__.__name__ in ['Conv1D', 'Dense']:
        w = layer.get_weights()[0]
        layersize = np.prod(w.shape)
        print("{}: {}".format(layer.name, layersize))  # 0 = weights, 1 = biases
        if layersize > 4096:  # assuming that shape[0] is batch, i.e., 'None'
            print("Layer {} is too large ({}), are you sure you want to train?".format(layer.name, layersize))
```

```python
import hls4ml
import os
os.environ['PATH'] = '/tools/Xilinx/Vitis_HLS/2023.2/bin:' + os.environ['PATH']
```

```python
hls_config = hls4ml.utils.config_from_keras_model(model, granularity='name')
hls_config['Model']['ReuseFactor'] = 1
hls_config['Model']['PipelineStyle'] = 'dataflow'



for Layer in hls_config['LayerName'].keys():
    hls_config['LayerName'][Layer]['Strategy'] = 'Resource'
    hls_config['LayerName'][Layer]['ReuseFactor'] = 5
    
hls_config['LayerName']['q_conv1d']['Strategy'] = 'Resource'
hls_config['LayerName']['q_conv1d']['ReuseFactor'] = 1
hls_config['LayerName']['q_conv1d']['Precision'] = 'ap_fixed<8,1>'

hls_config['LayerName']['q_activation']['Precision'] = 'ap_fixed<8,1>'

hls_config['LayerName']['q_conv1d_1']['Strategy'] = 'Resource'
hls_config['LayerName']['q_conv1d_1']['ReuseFactor'] = 150
hls_config['LayerName']['q_conv1d_1']['Precision'] = 'ap_fixed<8,1>'

hls_config['LayerName']['q_activation_1']['Precision'] = 'ap_fixed<8,1>'

hls_config['LayerName']['q_conv1d_2']['Strategy'] = 'Resource'
hls_config['LayerName']['q_conv1d_2']['ReuseFactor'] = 450
hls_config['LayerName']['q_conv1d_2']['Precision'] = 'ap_fixed<8,1>'

hls_config['LayerName']['q_activation_2']['Precision'] = 'ap_fixed<8,1>'

hls_config['LayerName']['q_dense']['Strategy'] = 'Resource'
hls_config['LayerName']['q_dense']['ReuseFactor'] = 111
hls_config['LayerName']['q_dense']['Precision'] = 'ap_fixed<8,1>'

print(hls_config)
```

```python
hls_model = hls4ml.converters.convert_from_keras_model(
    model,
    hls_config=hls_config,
    output_dir='cnn_fp8/',
    project_name='cnn_fp8',
    io_type='io_stream',
    backend='Vitis',
    part='xc7a200tfbg484-3' )

hls_model.compile()
```
